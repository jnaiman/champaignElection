```python
import geopandas
import matplotlib.pyplot as plt
%matplotlib inline
```

We will be using `geopandas` and the census webpage for 2019: https://www.census.gov/geographies/mapping-files/time-series/geo/tiger-line-file.html

Note that there is an FTP site, and it looks like they have info back until about 2010 and possibly back to 2008: https://www2.census.gov/geo/tiger/.  There might be other info back further too, its a little hard to tell.

```python
gdf = geopandas.read_file('/Users/jillnaiman/Dropbox/openChampaignProject/congressionalDistrictsTiger/tl_2019_us_cd116.shp')
```

```python
fig, ax = plt.subplots(figsize=(20,10))

gdf.plot(ax = ax)

ax.set_xlim(-180, -60)
ax.set_ylim(15, 75)

plt.show()
```
